```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
#converted values of 0 and 1 of SeniorCitizen as no and yes for better understanding
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

        
```

```python
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)

df.head()
```

```python
ax = sns.countplot(x = 'Churn' , data = df)

ax.bar_label(ax.containers[0])

plt.title("Count of customers by Churn")

plt.show()
```

```python
plt.figure(figsize = (3,4))

gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage Of Churned Customer")
plt.show()
```

```python
plt.figure(figsize = (4,3))

sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,3))

sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate counts
count_data = df.groupby(["SeniorCitizen", "Churn"]).size().unstack()
plt.figure(figsize = (4,4))
# Convert to percentage
count_percent = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
count_percent.plot(kind="bar", stacked=True, figsize=(6, 4), colormap="coolwarm")

# Add percentage labels
for i, bars in enumerate(plt.gca().containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Avoid labeling empty bars
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_y() + height/2, 
                     f"{height:.1f}%", ha="center", va="center", color="white", fontsize=10)

# Labels and title
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.title("Churn by Senior Citizen (Stacked %)")
plt.legend(title="Churn")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))
fig.suptitle("Customer Service Feature Distribution", fontsize=16)
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])  # ✅ Removed `palette`
    axes[i].set_title(f'{col}', fontsize=12)
    axes[i].tick_params(axis='x', rotation=15)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

#Customers using fiber optic internet and streaming services like StreamingTV and StreamingMovies 
#show higher churn, indicating a need for better service quality or bundling offers. Lack of security
#features such as OnlineSecurity and TechSupport is also linked to higher churn, suggesting these should
#be promoted or included in plans. Overall, improving service quality and value for customers 
#not using support features can help reduce churn.
